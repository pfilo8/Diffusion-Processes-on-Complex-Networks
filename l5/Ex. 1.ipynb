$$\frac{dS(t)}{dt} = bS - \beta IS $$

$$\frac{dI(t)}{dt} = \beta IS - kI $$

```python
import numpy as np
import matplotlib.pyplot as plt

from utils_solution import execute_experiment, generate_plot, plot_phase_plane
```

```python
def model_si(y, t, beta, b, k):
    St, It = y
    dS = b*St - beta*It*St
    dI = beta*It*St - k*It
    return [dS, dI]
```

```python
beta = 3
b = 3
k = 3
t = np.linspace(1, 10, 1001)

args = (beta, b, k)
s0 = 1
```

# Solution for $I(0) = 0.1$ 

```python
i0 = 0.1
solution = execute_experiment(model_si, [s0, i0], t, args)
generate_plot(t, solution, labels = ['S(t)', 'I(t)'])
```

# Solution for $I(0) = 0.3$ 

```python
i0 = 0.3
solution = execute_experiment(model_si, [s0, i0], t, args)
generate_plot(t, solution, labels = ['S(t)', 'I(t)'])
```

# Solution for $I(0) = 1.3$ 

```python
i0 = 1.3
solution = execute_experiment(model_si, [s0, i0], t, args)
generate_plot(t, solution, labels = ['S(t)', 'I(t)'])
```

# Phase plane

```python
fig, ax = plt.subplots(1, 1, figsize = (10, 8))

ax.set_xlabel('$S(t)$')
ax.set_ylabel('$I(t)$')

_ = plot_phase_plane(ax, model_si, (0, 6), (0, 6), args = args)
```
