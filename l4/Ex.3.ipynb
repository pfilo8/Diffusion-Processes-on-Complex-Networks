```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

from main3 import random_walk_on_graph, list_without_elements
```

```python
N = 100
MAX_STEPS = 10000
N_SIMULATIONS = 100

SOURCE_NODE = 0
TARGET_NODES = list_without_elements(N, SOURCE_NODE)
```

```python
def plot_graph(g, source_node):
    plt.figure(figsize = (12, 8))

    pos = nx.spring_layout(g)
    nx.draw_networkx(g, pos = pos)
    nx.draw_networkx_nodes(g, pos = pos, nodelist = [source_node], node_color = 'red')

def conduct_experiment(g):
    for node in TARGET_NODES:
        results = np.array([random_walk_on_graph(g, SOURCE_NODE, node) for _ in range(N_SIMULATIONS)])
        print(f'Path from {SOURCE_NODE} to {node} takes: {results[~np.isnan(results)].mean()} time steps.')
```

# Complete Graph

```python
g = nx.complete_graph(N)
plot_graph(g, SOURCE_NODE)
```

```python
conduct_experiment(g)
```

# Erdos Renyi Graph

```python
g = nx.erdos_renyi_graph(100, 0.3)
plot_graph(g, SOURCE_NODE)
```

```python
conduct_experiment(g)
```

# Watts Strogatz Graph

```python
g = nx.watts_strogatz_graph(100, 4, 0.3)
plot_graph(g, SOURCE_NODE)
```

```python
conduct_experiment(g)
```
